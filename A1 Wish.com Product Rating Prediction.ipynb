# In data mining, what is the difference between prediction and categorization?

Classification is about discovering a model that defines the data classes and concepts. The idea is to use this model to predict the class. The derived model is dependent on the examination of sets of training data.

Categorization is the exercise of creating meaningful categories for a particular variable/feature in the dataset, for better inference/understanding.Once we have few but meaningful categories, it becomes easier to do data analysis.

```python
#dropping unnecessary columns
df=df.drop(['merchant_has_profile_picture','has_urgency_banner','merchant_name','urgency_text','rating_count','shipping_option_name'],axis=1)
dft=dft.drop(['merchant_has_profile_picture','has_urgency_banner','merchant_name','urgency_text','rating_count','shipping_option_name'],axis=1)
df #viewing my training set after dropping them
```

```python
dft
```

```python
# I noticed that my data set has alot of categorical values 
#lots of machine learning algorithms won't be able to accept them so I'll need to turn them into numerical values
#to be honest I tried the hot encoding but I had a problem when I applied it seperatly on training set and testing set 
#as it gave me different number of columns for each set which is a super huge problem
#instead I decided to apply the label encoding and it worked very well!
#the following code is when I used the LabelEncoder to Encode Multiple Columns All at Once for both trainig and testing

#for training
from sklearn.preprocessing import LabelEncoder

cols = ["currency_buyer","tags","product_color","product_variation_size_id","merchant_title","merchant_info_subtitle","merchant_id","merchant_profile_picture","origin_country","theme","crawl_month"]

# Encode labels of multiple columns at once

df[cols] = df[cols].apply(LabelEncoder().fit_transform)

df
```

```python
#for testing
cols = ["currency_buyer","tags","product_color","product_variation_size_id","merchant_title","merchant_info_subtitle","merchant_id","merchant_profile_picture","origin_country","theme","crawl_month"]

# Encode labels of multiple columns at once

dft[cols] = dft[cols].apply(LabelEncoder().fit_transform)

dft
```

```python
#Splitting the data into features and target variables
xtr=df.loc[:, df.columns != 'rating'] #here I started splitting my training set (df) into x (input)
ytr=df.rating #and y the output (my target) which tells the rating from 1 to 5
xts=dft.drop(['id'],axis=1) #on my testing data, I don't wanna consider the id column as a feature so i dropped it
idts=dft['id'] #and here i put the dropped id column in another variable cause I will need in the end to call the product with its id while testing
xts #just vieweing my testing dataset
```

```python
xtr
```

```python
#Splitting training data into 80% Train , 20% Validation
from sklearn.model_selection import train_test_split
xtr_train, xtr_valid, ytr_train, ytr_valid = train_test_split(xtr, ytr, test_size = 0.2, random_state=3)
xtr_train
```

```python
ytr_train
```

```python
#let's start applying our machine learning algorothim and I wanted to begin with the decision tree model
from sklearn.tree import DecisionTreeClassifier
# I've created the decision tree classifier by passing the  random state paramter to DecisionTreeClassifier().
model = DecisionTreeClassifier(random_state=3)
#the training process by using the model.fit() method. (before tuning)
model.fit(xtr_train, ytr_train)
y_pred = model.predict(xtr_valid) 
y_pred #showing my predictions before tuning
```

```python
#Tuning the hyperparameters using randomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
# Setup the parameters and distributions to sample from: param_dist
param_dist = {"max_depth": [3, None],
              "max_features": randint(1, 9),
              "min_samples_leaf": randint(1, 9),
              "criterion": ["gini", "entropy"]}
# Instantiate the RandomizedSearchCV object: tree_cv
tree_cv = RandomizedSearchCV(model, param_dist, cv=5)
# Fit it to the data
tree_cv.fit(xtr_train,ytr_train)
```

```python
#Validation after tuning
y_pred = tree_cv.predict(xtr_valid)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(ytr_valid, y_pred))
print(classification_report(ytr_valid, y_pred))
```

```python
from sklearn.metrics import f1_score
f1_score(ytr_valid, y_pred, average='micro') 
```

```python
# concatenating x_test and id_test along columns in order to be able to call products with its ids when testing
final_test = pd.concat([xts,idts], axis=1)
final_test.set_index("id", inplace = True)
```

```python
_id = final_test.index
```

```python
#Testing Decision Tree

pred_test = tree_cv.predict(final_test) #storing our predictions is pred_test file

pred_df = pd.DataFrame(data={'id': np.asarray(_id), 'rating': pred_test})

pred_df.to_csv('pred_walkthrough.csv', index=False) #generating the submission file
```

```python
#I need to scale my data before applying the svm model, I will use MinMaxScaler()
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
xtr_train=pd.DataFrame(scaler.fit_transform(xtr_train),
            columns=xtr_train.columns, index=xtr_train.index) #scaling the train set
xtr_train #viewing it after scaling


```

```python
#Now the svm model
from sklearn.svm import SVC
model = SVC() #creating the svm model
model.fit(xtr_train, ytr_train) #fitting the model
y_pred = model.predict(xtr_valid) 
y_pred #showing my predictions before tuning
```

```python
#tuning the parameters using GridSearchCV
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],'gamma': [1, 0.1, 0.01, 0.001, 0.0001],'kernel': ['rbf']}

grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3)

# fitting the model for grid search
grid.fit(xtr_train, ytr_train)
```

```python
# print best parameter after tuning
print(grid.best_params_)
 
# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
#Validation after tuning
y_pred = grid.predict(xtr_valid)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(ytr_valid, y_pred))
print(classification_report(ytr_valid, y_pred))
```
